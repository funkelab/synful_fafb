```python
from credentials import credentials
import pymaid
from tqdm import tqdm
import numpy as np
import pandas as pd
from pymongo import MongoClient
from matplotlib import pyplot as plt


input_site_color ='#72b9cb'
output_site_color = '#c12430'
# output_site_color= 'blue'


def _connect_mongo(host, port, username, password, db):
    """ A util for making a connection to mongo """

    if username and password:
        mongo_uri = 'mongodb://%s:%s@%s:%s/%s' % (username, password, host, port, db)
        conn = MongoClient(mongo_uri)
    else:
        conn = MongoClient(host, port)


    return conn[db]


def read_mongo(db, collection, query={}, host='localhost', port=27017, username=None, password=None, no_id=True):
    """ Read from Mongo and Store into DataFrame """

    # Connect to MongoDB
    db = _connect_mongo(host=host, port=port, username=username, password=password, db=db)

    # Make a query to the specific DB and Collection
    cursor = db[collection].find(query)

    # Expand the cursor and construct the DataFrame
    df =  pd.DataFrame(list(cursor))

    # Delete the _id
    if no_id:
        del df['_id']

    return db, df

def get_ann_by_vol(
        annotations,
        neuropils=None,
        skids=None,):

    '''

    Fetch neurons by annotation(s) from catmaid. Prune by intersection with
    neuropil(s) if requested

    Args:

        annotations (``string`` or ``list of strings``):

            The neuron annotation or annotations to query

        neuropils (``string`` or ``list of strings``):

            The neuropils in which to filter the annotations (optional). If
            passing multiple neuropils use a list of strings, the volumes will
            be joined before intersection

    '''

    # fetch skeletons
    if not skids:
        skids = pymaid.get_skids_by_annotation(annotations)
    print("Loading skeleton ids: ", skids)
    # convert to 3d neurons

    neurons = pymaid.get_neurons(skids)

    # get neuropil(s)

    vol = pymaid.get_volume(neuropils)

    # if multiple neuropils requested, combine into single mesh first

    if type(neuropils) == list and len(neuropils) > 1:
        print('Multiple neuropils requested to filter, merging into one volume...')
        vol = pymaid.Volume.combine(vol)

    # prune neurons by volume if requesting neuropils

    if neuropils:
        print('Filtering id(s): %s by volume(s): %s' %(skids, neuropils))
        neurons = neurons.prune_by_volume(vol, inplace=False)

    return neurons

def plot_neuron(neuron):
    return neuron.plot2d(cn_size=15, connectors=False, scalebar=10, color='black')

def plot_synapses(x, y, ax, color='red', s=10):
    ax.scatter(x,
           -y,
           c=color, alpha=1.0, zorder=4,
           edgecolor='none', s=s)
    

def plot_gt(neuron, config_eval, skel_id, dataset, input_synapses=False, output_synapses=False, s=10):
    gt_db_host = config_eval['gt_db_host']
    gt_db_name = config_eval['gt_db_name']
    gt_db_col = config_eval['gt_db_col'] + '.synapses'
    db_syn_gt, df_syn_gt = read_mongo(gt_db_name, gt_db_col, host=gt_db_host)
    
    fig, ax = plot_neuron(neuron)
    connector_shift = 0.6
    if output_synapses:
        df_single_gt = df_syn_gt[df_syn_gt.pre_skel_id == skel_id]
#         plot_synapses(df_single_gt.post_x, df_single_gt.post_y, ax, color=output_site_color, s=s)
        plot_synapses(df_single_gt.pre_x+connector_shift*(df_single_gt.post_x-df_single_gt.pre_x), df_single_gt.pre_y+connector_shift*(df_single_gt.post_y-df_single_gt.pre_y), ax, color=output_site_color, s=s)
    if input_synapses:
        df_single_gt = df_syn_gt[df_syn_gt.post_skel_id == skel_id]
        plot_synapses(df_single_gt.post_x, df_single_gt.post_y, ax, color=input_site_color, s=s)
    outputfile = f'{dataset}_gt.csv'
    df_single_gt.to_csv(outputfile, index=False)
            
def plot_pred(neuron, config_eval, skel_id, dataset, input_synapses=False, output_synapses=False, score_thr=0, s=10):
#     fig = plt.figure((10, 10))
    
    db_host = config_eval['res_db_host']
    db_name = config_eval['res_db_name']
    db_col = config_eval['res_db_col'] + f".syn_thr{int(score_thr) * 1000}.synapses"

    db_syn_pred, df_syn_pred = read_mongo(db_name, db_col, host=db_host)
    
    
    fig, ax = plot_neuron(neuron)
    if output_synapses:
        df_single = df_syn_pred[df_syn_pred.pre_skel_id == skel_id]
        plot_synapses(df_single.pre_x, df_single.pre_y, ax, color=output_site_color, s=s)
    if input_synapses:
        df_single = df_syn_pred[df_syn_pred.post_skel_id == skel_id]
        plot_synapses(df_single.post_x, df_single.post_y, ax, color=input_site_color, s=s)
    
    outputfile = f'{dataset}_pred.csv'
    df_single.to_csv(outputfile, index=False)
    res_db_host = config_eval['res_db_host']
    res_db_name = config_eval['res_db_name']
    res_db_col = config_eval['res_db_col'] + f".thr{int(score_thr)*1000}"
    client_res = MongoClient(res_db_host)
    db_res = client_res[res_db_name]
    col = db_res[res_db_col]
    summary = db_res['summary']
    summary = pd.DataFrame(summary.find({'score_thr': score_thr}))

    res = list(col.find({'skel_id': skel_id}))[0]
    print('fscore: {}'.format(res['fscore']))
    print(f'score threshold:{score_thr}')
```
